```python
#rug plot and a historgram for a dataset
dataset = randn(25)
sns.rugplot(dataset)

plt.hist(dataset, alpha = 0.5)
```

```python
#kde with a dataset in seaborg
sns.kdeplot(dataset)
```

```python
#to adjust bandwidth, specifying it
sns.rugplot(dataset, color = 'black')
for bw in np.arange(0.5, 2, 0.25):
    sns.kdeplot(dataset, bw = bw, lw = 1.8, label = bw)
```

```python
#Kern plot in different kernel types like gaussion
kernel_options = ['biw', 'epa', 'gau', 'tri', 'triw']
for kern in kernel_options: 
    sns.kdeplot(dataset, kernel = kern, label = kern, shade = True )
```

```python
sns.kdeplot(dataset, cumulative = True)
```

```python
#Multivariate normal distribution, with more than 2 variables
mean = [0,0]
cov = [[1,0],[0,100]]
dataset2 = np.random.multivariate_normal(mean, cov, 1000) 

dframe = pd.DataFrame(dataset2, columns =['X', 'Y'])
sns.kdeplot(dframe, shade = True)
#To get a joint plot 
sns.jointplot('X','Y',dframe, kind = 'kde')
```
